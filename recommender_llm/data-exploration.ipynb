```python
from statistics import correlation

import kagglehub
from seaborn import heatmap
from sqlalchemy.dialects.mssql.information_schema import columns

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books=pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax=plt.axes(
)
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)
plt.xlabel("Columns")
plt.ylabel("Missing Values")
plt.show()
```

```python
import numpy as np

books["missing_description"]=np.where(books["description"].isna(),1,0)
books["age_of_book"]=2025-books["published_year"]
```

```python
columns_of_interest=["num_pages","age_of_book","missing_description","average_rating"]

correlation_matrix=books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap=sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",fmt=".2f",cbar_kws={"label": "Spearmen Correlation"})
heatmap.set_title("Correlation Matrix")
plt.show()
```

```python
book_missing=books[~(books["description"].isna()) & ~(books["published_year"].isna()) &
      ~(books["average_rating"].isna() )& ~(books["num_pages"].isna())]
```

```python
book_missing_25_words=book_missing[book_missing["words_in_description"]>=25]
```

```python
book_missing_25_words["title_and_subtitle"]=(
    np.where(book_missing_25_words["subtitle"].isna(),book_missing_25_words["title"],book_missing_25_words[["title","subtitle"]].astype(str).agg(": ".join,axis=1))
)
```

```python
book_missing_25_words["tagged_description"]=book_missing_25_words[["isbn13","description"]].astype(str).agg(" ".join,axis=1)
```
